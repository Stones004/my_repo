```python
import os
import json
from mineru_vl_utils import MinerUClient
from pdf2image import convert_from_path

INPUT_FILE = "/content/drive/MyDrive/MinerU_input/Netweb.pdf"
OUTPUT_FILE = "/content/drive/MyDrive/MinerU_output/test_output.json"
MODEL_PATH = "/content/drive/MyDrive/MinerU_model"

print("Loading MinerU...")
client = MinerUClient(
    backend="transformers",
    model_path=MODEL_PATH
)
print("Model Loaded.\n")

print("Converting PDF to images...")
images = convert_from_path(INPUT_FILE, dpi=200)

results = []

for i, image in enumerate(images, start=1):
    print(f"Processing page {i}...")
    output = client.two_step_extract(image)
    results.append({
        "page": i,
        "output": output
    })

with open(OUTPUT_FILE, "w") as f:
    json.dump(results, f)

print("\nDone.")
print("Saved to:", OUTPUT_FILE)
```
